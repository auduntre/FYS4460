```python
rdf = [line.split(" ")[:] for line in open("rdf_0.1.txt")][3:]
rdf = array([array(line, dtype=float) for line in rdf])
N = 20
```

```python
mask = np.ones(rdf.shape[0], dtype=bool)
mask[0::N+1] = False

rdf = rdf[mask]
rdf = stack(rdf)
```

```python
rdf.shape
```

```python
figure(figsize=(10, 5))
plot(rdf[:N, 1], rdf[20000:20000+N, 2])
show()
```
