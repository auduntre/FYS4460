```python
from pylab import *
```

```python
%matplotlib inline
```

```python
rcParams['font.size'] = 14 
```

```python
ts = [float(str(0.1 * i)) for i in range(1, 41)]

temps = zeros(len(ts))
D = zeros(len(ts))
dt = 0.005
numb_time_steps = 10000
```

```python
for i, ti in enumerate(ts):
    r_msd = (loadtxt("msd_{}.txt".format(ti))[-1, 1])
    D[i] = r_msd / (6.0 * numb_time_steps  * dt)
    _, TEMP = loadtxt("msd_t{}.txt".format(ti)).T
    temps[i] = TEMP[-1]
    
sort_idx = argsort(temps)
```

```python
figure(figsize=(10, 5))
ylim(0, 0.03)
xlim(0, 1.)
plot(temps[sort_idx], D[sort_idx])
title("D(t_end) (rho=0.8, solid-liquid)")
xlabel("Temperature")
ylabel("D")
show()
```
