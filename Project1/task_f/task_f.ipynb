```python
ts = ["{:.2f}".format(0.05*i) for i in range(81)]

temps = zeros(len(ts))
D = zeros(len(ts))
dt = 0.0025
```

```python
for i, ti in enumerate(ts):
    _, TEMP = loadtxt("msd_t{}.txt".format(ti)).T
    r_msd = loadtxt("msd_{}.txt".format(ti)[-1, 1]
    temps[i] = TEMP[-1]
    D[i] = r_msd / (6.0 * len(TEMP) * dt)
    
sort_idx = argsort(temps)
```

```python
figure(figsize=(10, 5))
#ylim(0, 0.03)
#xlim(0, 1.)
plot(temps[sort_idx], D[sort_idx])
title("D(t_end) (rho=0.8, solid-liquid) with bath")
xlabel("Temperature")
ylabel("D")
show()
```
