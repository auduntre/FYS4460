```python
from pylab import *
```

```python
%matplotlib inline
```

```python
ts = [float(str(0.1 * i)) for i in range(1, 41)]

temps = []
D = []
dt = 0.005
numb_time_steps = 10000
```

```python
for ti in ts:
    r_msd = loadtxt("msd_{}.txt".format(ti))[-1, 1]
    D.append(r_msd / (6.0 * numb_time_steps  * dt))
    _, TEMP = loadtxt("msd_t{}.txt".format(ti)).T
    temps.append(TEMP[-1])
```

```python
figure(figsize=(10, 5))
#ylim(1.1, 1.5)
#xlim(3, 4)
plot(temps, D)
xlabel("temp")
ylabel("D")
show()
```
