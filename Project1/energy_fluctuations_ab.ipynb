```python
data_smallest = dump("dump.smallesttimestep")
data_small = dump("dump.smalltimestep")
data = dump("dump.lammpstrj")
data_medium = dump("dump.mediumtimstep")
data_big = dump("dump.bigtimestep")
```

```python
%matplotlib inline
```

```python
rcParams['figure.figsize'] = [16, 8]
rcParams['font.size'] = 16
```

```python
t = [data_smallest.time(), data_small.time(), data.time(), data_medium.time(), data_big.time()]
step_size = [0.00125, 0.0025, 0.005, 0.008,  0.01]
data_sets = ["smallesttimestep.txt","smalltimestep.txt", "compute.txt", 
             "mediumtimestep.txt", "bistimestep.txt"]

times = []

for i, t_ in enumerate(t):
    times.append([step_size[i]*ti for ti in t_])
```

```python
total_energy_sets = []

for ds in data_sets:
    _, KE, PE, _, _ = loadtxt(ds).T
    total_energy_sets.append(KE + PE)
```

```python
for i, times_ in enumerate(times):
    plot(times_, total_energy_sets[i][t[i]], label="timestep size = {}".format(step_size[i]))

title("Total energy for different sized time-steps")
xlabel("Time [lj]")
ylabel("Total energy [lj]")
legend()
show()
```

```python
for i, te in enumerate(total_energy_sets):
    print("With step size {} variance = {}, std = {}".format(step_size[i], te.var(), te.std()))
```

```python
for i, times_ in enumerate(times):
    cutoff = int(len(t[i])*0.1)
    plot(times_[cutoff:], total_energy_sets[i][t[i]][cutoff:], label="timestep size = {}".format(step_size[i]))

title("Total energy for different sized time-steps")
xlabel("Time [lj]")
ylabel("Total energy [lj]")
legend()
show()
```

```python
for i, te in enumerate(total_energy_sets):
    cutoff = int(len(t[i])*0.1)
    print("With step size {} variance = {}, std = {}".format(step_size[i], te[cutoff:].var(), te[cutoff:].std()))
```
