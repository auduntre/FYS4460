```python
from pylab import *
import seaborn
seaborn.set()

%matplotlib inline
```

```python

N = 28
D = zeros(N)
temps = zeros(N)

dt = 0.005
timesteps = 10000
```

```python
for i in range(N):
    numb = 1000 + 500*i
    r_msd = (loadtxt("msd_si_{}.txt".format(numb))[4::5, 1])[5000:].mean()
    D[i] = r_msd / (6.0 * dt * timesteps)
    temps[i] = loadtxt("compute_si_{}.txt".format(numb)).T[1, 5000:].mean()
```

```python
sort_idx = argsort(temps)

figure(figsize=(10,6))
plot(temps[sort_idx], D[sort_idx])
axvline(x=1687, linestyle='dashed', color='red')

title("Detection of phase change by studying D(T) for silicon")
xlabel("Temperature")
ylabel("D(T)")
show()
```
